```python
numpyArray1=np.linspace(1,10,20)
```

```python
numpyArray1
```

```python
numpyArray2=numpyArray1**3
```

```python
plt.plot(numpyArray1,numpyArray2,"b*-")
```

```python
plt.subplot(1,2,1)##Bir sÄ±ra olacak ve 2 grafik olacak . 1. grafikteyim
plt.plot(numpyArray1,numpyArray2,"r--"),
plt.xlabel("SayÄ±")
plt.ylabel("KÃ¼pÃ¼")
plt.subplot(1,2,2)##Bir sÄ±ra olacak 2 grafik olacak ve 2. grafikteyim
plt.plot(numpyArray2,numpyArray1,"b*-")
plt.xlabel("KÃ¼pÃ¼")
plt.ylabel("SayÄ±")

plt.tight_layout()
```

```python
benimFigure=plt.figure()
figureAxces=benimFigure.add_axes([0.2,0.2,0.5,0.5])
figureAxces.plot(numpyArray1,numpyArray2,"b--")
figureAxces.set_xlabel("sayÄ±")
figureAxces.set_ylabel("kup")
figureAxces.set_title("SayÄ± KÃ¼p GrafiÄŸi")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])


eksen1.plot(numpyArray1,numpyArray2,"b")
eksen1.set_title("eksen1 grafiÄŸi")
eksen1.set_xlabel("eksem1 X")
eksen1.set_ylabel("eksen1 Y")

eksen2.plot(numpyArray2,numpyArray1,"r")
eksen2.set_title("eksen2 grafiÄŸi")
eksen2.set_xlabel("eksem2 X")
eksen2.set_ylabel("eksen2 Y")
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyArray1,numpyArray2,"r")
    eksen.set_xlabel("X")
    
plt.tight_layout()## grafikler arasÄ±nda mesafeyi aÃ§Ä±p daha gÃ¼zel gÃ¶rÃ¼nÃ¼m
#benimFigure
```

```python
yeniFigure=plt.figure()
#dpi=150

yeniEksen=yeniFigure.add_axes([0.2,0.2,0.8,0.8])

yeniEksen.plot(numpyArray1,numpyArray1**2,label="sayÄ±nÄ± karesi")
yeniEksen.plot(numpyArray1,numpyArray1**3,label="sayÄ±nÄ±n kÃ¼pÃ¼")
yeniEksen.legend(loc=6)##loc yerini belirtiyo
```
