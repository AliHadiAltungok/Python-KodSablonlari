```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpyDizi1=np.linspace(1,10,20)
```

```python
numpyDizi1
```

```python
numpyDizi2=numpyDizi1**2
```

```python
(benimFigur,benimEksen)=plt.subplots()

benimEksen.plot(numpyDizi1,numpyDizi2,color="#DF27E9",alpha=0.7)
benimEksen.plot(numpyDizi2,numpyDizi1,"#841212",alpha=0.3)
```

```python
(yeniFigure,yeniEksen)=plt.subplots()
yeniEksen.plot(numpyDizi1,numpyDizi1+2,color="yellow",linewidth=1.0)
yeniEksen.plot(numpyDizi1,numpyDizi1+3,color="#84D212",linewidth=5.0)
yeniEksen.plot(numpyDizi1,numpyDizi1+4,color="#84D287",linestyle="-.")
yeniEksen.plot(numpyDizi1,numpyDizi1+5,color="#844FC5",linestyle=":")
yeniEksen.plot(numpyDizi1,numpyDizi1+6,color="#EF8F00",linestyle="--")

yeniEksen.plot(numpyDizi1,numpyDizi1+8,color="#000000",linestyle="--",marker="+",markersize=8)

yeniEksen.plot(numpyDizi1,numpyDizi1+10,color="#000000",linestyle="--",marker="o",markersize=8,markerfacecolor="red")
```
